```python
# Produce the Summaries using T5
import os, glob, re, nltk, json
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

def clean_line(line):
    line = line.lower()
    line = re.sub(r'[^\w\s]', '', line)
    line = re.sub(r'\s+', ' ', line)
    return line

def load_data(data_dir):
    """Load .txt.data files into dict: topic -> list of cleaned review lines"""
    data = {}
    search_path = os.path.join(data_dir, "*.txt.data")
    files = glob.glob(search_path)
    for file in files:
        filename = os.path.basename(file)
        base = os.path.splitext(filename)[0]
        topic = os.path.splitext(base)[0]
        with open(file, 'r', encoding='utf-8', errors='ignore') as f:
            lines = []
            for line in f:
                if line.strip():
                    lines.append(clean_line(line.strip()))
            data[topic] = lines
    return data

model_name = "t5-small"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()

def summarize_text(text, max_length=100):
    """Generate a summary from text using T5"""
    prefix = "summarize: " + text
    inputs = tokenizer(prefix, max_length=512, truncation=True, return_tensors="pt").to(device)
    with torch.no_grad():
        summary_ids = model.generate(
            inputs["input_ids"],
            attention_mask=inputs["attention_mask"],
            max_length=max_length,
            num_beams=4,
            early_stopping=True
        )
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary.strip()

def summarize_all_topics(data_dir):
    data = load_data(data_dir)
    print(f"Loaded topics: {list(data.keys())}")
    summaries = {}

    for topic, reviews in data.items():
        # Join all reviews for the topic into one string for summarization
        full_text = " ".join(reviews)
        print(f"Summarizing topic: {topic} (length: {len(full_text)} chars)")
        summary = summarize_text(full_text)
        summaries[topic] = [summary]

    # Save summaries to JSON file
    with open("t5_summaries.json", "w", encoding="utf-8") as f:
        json.dump(summaries, f, indent=2, ensure_ascii=False)
    print("Saved summaries to t5_summaries.json")

if __name__ == "__main__":
    nltk.download('punkt')  # just in case
    summarize_all_topics("data")
```

```python
# 6 Evaluate Predicted Summaries Using HuggingFace-style ROUGE
import json
import os
import numpy as np
from nltk import sent_tokenize
from rouge_score import rouge_scorer

# Load generated summaries from file (json format: {topic: [summary_text]})
with open("t5_summaries.json", "r", encoding="utf-8") as f:
    pred_dict = json.load(f)

# ??? Insert sentence boundaries (for better ROUGE scoring)
def insert_boundaries(text):
    sentences = sent_tokenize(text)
    return "\n".join(sentences)

predictions, references = [], []

for topic, pred_list in pred_dict.items():
    pred = pred_list[0].strip()
    predictions.append(pred)

    gold_path = os.path.join("summaries-gold", topic, f"{topic}.1.gold")
    try:
        with open(gold_path, "r", encoding="utf-8") as gf:
            gold = gf.read().strip()
        references.append(gold)
    except FileNotFoundError:
        print(f" Error: Missing gold summary for topic: {topic}")
        predictions.pop()  # drop prediction if no reference
        continue

# initialize scorer
scorer = rouge_scorer.RougeScorer(["rouge1", "rouge2", "rougeL"], use_stemmer=True)

#compute ROUGE scores
rouge1_p, rouge1_r, rouge1_f = [], [], []
rouge2_p, rouge2_r, rouge2_f = [], [], []
rougeL_p, rougeL_r, rougeL_f = [], [], []

for pred, ref in zip(predictions, references):
    scores = scorer.score(insert_boundaries(pred), insert_boundaries(ref))
    rouge1_p.append(scores["rouge1"].precision)
    rouge1_r.append(scores["rouge1"].recall)
    rouge1_f.append(scores["rouge1"].fmeasure)

    rouge2_p.append(scores["rouge2"].precision)
    rouge2_r.append(scores["rouge2"].recall)
    rouge2_f.append(scores["rouge2"].fmeasure)

    rougeL_p.append(scores["rougeL"].precision)
    rougeL_r.append(scores["rougeL"].recall)
    rougeL_f.append(scores["rougeL"].fmeasure)

# Output average scores
huggingface_rouge_scores = {
    "rouge1_precision": round(np.mean(rouge1_p), 4),
    "rouge1_recall": round(np.mean(rouge1_r), 4),
    "rouge1_f1": round(np.mean(rouge1_f), 4),
    "rouge2_precision": round(np.mean(rouge2_p), 4),
    "rouge2_recall": round(np.mean(rouge2_r), 4),
    "rouge2_f1": round(np.mean(rouge2_f), 4),
    "rougeL_precision": round(np.mean(rougeL_p), 4),
    "rougeL_recall": round(np.mean(rougeL_r), 4),
    "rougeL_f1": round(np.mean(rougeL_f), 4),
}

print("\n--- Final ROUGE Scores ---")
for k, v in huggingface_rouge_scores.items():
    print(f"{k}: {v}")
```

```python
# 6A. Evaluate model using HuggingFace
!pip install -q rouge-score evaluate

import os
import json
from nltk import sent_tokenize, download
import evaluate

download("punkt")
download("punkt_tab")

def insert_boundaries(text):
    return "\n".join(sent_tokenize(text))

with open("t5_summaries.json", "r", encoding="utf-8") as f:
    system_outputs = json.load(f)  # dict: topic -> [summary]

# Load gold summaries from summaries-gold/
gold_summaries = {}
for topic in system_outputs:
    topic_dir = os.path.join("summaries-gold", topic)
    gold_text = ""
    for filename in os.listdir(topic_dir):
        if filename.endswith(".gold"):
            with open(os.path.join(topic_dir, filename), "r", encoding="utf-8") as f:
                gold_text += f.read().strip() + " "  # concatenate all gold summaries for topic
    gold_summaries[topic] = gold_text.strip()

preds = []
refs = []

for topic in system_outputs:
    sys_sum = system_outputs[topic][0]
    gold_sum = gold_summaries.get(topic, "")

    # Insert sentence boundaries for ROUGE-Lsum compatibility
    preds.append(insert_boundaries(sys_sum))
    refs.append(insert_boundaries(gold_sum))

# compute ROUGE
rouge = evaluate.load("rouge")
results = rouge.compute(predictions=preds, references=refs, use_stemmer=True)

#display results
print("ROUGE Results:")
for k, v in results.items():
    print(f"{k}: {v:.4f}")
```

```python
# 6B. Alternative ROUGE evaluation using rouge_metric PyRouge
!pip install -q rouge-metric

import os
import json
import re
from nltk.stem import PorterStemmer
from nltk import download
from rouge_metric import PyRouge

download("punkt")

stemmer = PorterStemmer()
rouge = PyRouge(
    rouge_n=(1, 2, 4),
    rouge_l=True,
    rouge_w=True,
    rouge_s=True,
    rouge_su=True,
    skip_gap=4
)

# minimal preprocessing
def preprocess(text):
    return text.strip().split()

# Load model pred summaries
with open("t5_summaries.json", "r", encoding="utf-8") as f:
    system_outputs = json.load(f)  # dict: topic -> [summary]

# Load gold summaries
gold_summaries = {}
for topic in system_outputs:
    topic_dir = os.path.join("summaries-gold", topic)
    gold_text = ""
    for filename in os.listdir(topic_dir):
        if filename.endswith(".gold"):
            with open(os.path.join(topic_dir, filename), "r", encoding="utf-8") as f:
                gold_text += f.read().strip() + " "
    gold_summaries[topic] = gold_text.strip()

hypotheses = []
references = []

for topic in system_outputs:
    pred = system_outputs[topic][0]
    ref = gold_summaries.get(topic, "")

    hypotheses.append(preprocess(pred))
    references.append([preprocess(ref)])  # reference must be a list of strings

#compute scores
scores = rouge.evaluate_tokenized(hypotheses, references)

print("ROUGE-1:", scores["rouge-1"])
print("ROUGE-L:", scores["rouge-l"])


# Print an example
print("\n--- ORIGINAL ---")
print("Prediction:", hypotheses[0])
print("Gold:", references[0][0])

# Print processed example
print("\n--- PyROUGE-style Preprocessed ---")
print("Pred:", (hypotheses[0]))
print("Gold:", (references[0][0]))

# Return score dict
results = {
    "rouge1_precision": round(scores["rouge-1"]["p"], 4),
    "rouge1_recall": round(scores["rouge-1"]["r"], 4),
    "rouge1_f1": round(scores["rouge-1"]["f"], 4),
    "rouge2_precision": round(scores["rouge-2"]["p"], 4),
    "rouge2_recall": round(scores["rouge-2"]["r"], 4),
    "rouge2_f1": round(scores["rouge-2"]["f"], 4),
    "rougeL_precision": round(scores["rouge-l"]["p"], 4),
    "rougeL_recall": round(scores["rouge-l"]["r"], 4),
    "rougeL_f1": round(scores["rouge-l"]["f"], 4),
    "rougeSU4_precision": round(scores["rouge-su4"]["p"], 4),
    "rougeSU4_recall": round(scores["rouge-su4"]["r"], 4),
    "rougeSU4_f1": round(scores["rouge-su4"]["f"], 4),
}

# Show final scores
print("\nFinal PyROUGE Scores:")
for k, v in results.items():
    print(f"{k}: {v}")
```
